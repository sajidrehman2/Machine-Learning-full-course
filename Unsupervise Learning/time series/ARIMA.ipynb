## ARIMA Model Building

#### Arima model has three components 
- 1. AR
- 2. MA
- 3. Integerated

For building ARIMA model we need the 
- AR order p
- MA order q
- Integerated order d

### Hyperparameters of ARIMA model (p,d,q)

```python
# p=7, d=0,q=1
```

```python
# How to get pdq, 3 different methods discussed below
```

## Build AIRMA model

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model  = ARIMA(train,order=(7,0,1)).fit()
```

### Preddict on Test Data

```python
len(ts2)
```

```python
pred = model.predict(start=120,end=141)
pred
```

## Model EVALUATION

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(pred,test))
```

```python
test.mean(), np.sqrt(test.var())
```

#### How to get ARIMA Hyperparameters pdq 

```python
import itertools
```

```python
p=range(0,8)
d=range(0,8)
q= range(0,2)
```

```python
list(p), list(d), list(q)
```

```python
pdq_combinations  = list(itertools.product(p,d,q))
```

```python
pdq_combinations
```

```python
rmse=[]
order1 = []
import warnings
warnings.filterwarnings('ignore')
```

```python
for pdq in pdq_combinations:    
    model=ARIMA(train,order=pdq).fit()
    pred = model.predict(start=120,end=141)
    error = np.sqrt(mean_squared_error(test,pred))
    print(error)
    order1.append(pdq)
    rmse.append(error)
```

```python
result = pd.DataFrame(index=order1,data=rmse,columns=['rmse'])
result
```
