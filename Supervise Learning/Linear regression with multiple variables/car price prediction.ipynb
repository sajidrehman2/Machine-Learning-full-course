```python
X=df[['enginesize', 'curbweight', 'horsepower', 'citympg', 'highwaympg', 'carwidth', 'carlength', 'wheelbase']]

y=df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
cdf=pd.DataFrame(lm.coef_,X.columns,columns=['Coef']) # the lenght of member is high correlated with yearly amount spend
print(cdf)
```

```python
predictions=lm.predict(X_test)
```

```python
# Plot actual values
sns.scatterplot(x=y_test, y=y_test, color='green', label='Actual Values')

# Plot predicted values
sns.scatterplot(x=predictions, y=y_test, color='red', label='Predicted Values')

# Add labels and title
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.title("Prediction vs Price")
plt.legend()  # Add a legend to differentiate between the actual and predicted values

plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
print('Mean Absolute Error:',mean_absolute_error(y_test, predictions))
print('Mean Squared Error:',mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, predictions)))
```

```python
residuals = y_test-predictions
sns.distplot(residuals, bins=30)
```

```python
import pylab 
import scipy.stats as stats

stats.probplot(residuals, dist="norm", plot=pylab)
pylab.show()
```

# Summary:
#### Most Positive Impact:
- enginesize has the highest positive correlation (0.874), indicating it has the strongest positive effect on car prices.
#### Most Negative Impact:
- highwaympg has the highest negative correlation (-0.697), suggesting that higher highway miles per gallon are associated with lower car prices.
# Interpretation:
#### Positive Variables:
-Higher values in enginesize, curbweight, horsepower, carwidth, carlength, wheelbase, and boreratio are likely to lead to higher car prices.
#### Negative Variables:
-Higher values in citympg and highwaympg may indicate a less expensive car, potentially because they represent fuel efficiency.
This analysis can help the car company understand which features to focus on when entering a new market or designing new models.
